```python
from PIL import Image

f = open('annotate.data',"r")
image_name=[]
xmin=[]
ymin=[]
width=[]
height=[]
width_1=[]
height_1=[]

for line in f.readlines():
    a=line.split("\\")
    b=a[1].split(" ")
    if(b[1]=="1"):
        try:
            im = Image.open('images/'+b[0])
            Width, Height = im.size
            width_1.append(Width)
            height_1.append(Height)
        
        except FileNotFoundError:
            width_1.append(0)
            height_1.append(0)
            
        
        image_name.append(b[0])
        xmin.append(b[2])
        ymin.append(b[3])
        width.append(b[4])
        height.append(b[5].replace("\n",""))
        
   
print(image_name,xmin,ymin,width,height)             
f.close()
```

```python
import csv
 
# myData = [image_name,width,height,class_name,xmin,ymin,xmax,ymax]
with open('final.csv', 'w') as csvfile:
    fieldnames = ['image_name','width','height','class','xmin','ymin','xmax','ymax']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
 
    writer.writeheader()
    for i in range(len(image_name)):
        writer.writerow({'image_name': image_name[i], 'width': width_1[i], 'height': height_1[i],
                         'class':'vest','xmin':xmin[i],'ymin':ymin[i],'xmax':int(xmin[i])+int(width[i]),
                          'ymax':int(ymin[i])+int(height[i])})
   
     
print("Writing complete")
```
